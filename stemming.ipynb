# Stemmer

```python
## classification problem
## comments of product is a positive review or negative review 
## Reviews ---> eatning, eat, eaten [going, gone, goes]
# https://github.com/krishnaik06/The-Grand-Complete-Data-Science-Materials
words = [
   'eating','eats','writing','writes','programming','history','finally','finalized']

```

# PorterStemmer

```python
from nltk.stem import PorterStemmer 
```

```python
stemming=PorterStemmer()
```

```python
for word in words: 
    print(word+"------>"+stemming.stem(word))
```

```python
stemming.stem('congratulation')
```

```python
stemming.stem('modification')
```

```python
stemming.stem('sitting')
```

# RegexpStemmer Class

```python
from nltk.stem import RegexpStemmer 
reg_stemmer=RegexpStemmer('ing$|s$|e$|able$', min=4)
```

```python
reg_stemmer.stem('eating')
```

```python
reg_stemmer.stem('ingeating')
```

# Snowball Stemmer

```python
from nltk.stem import SnowballStemmer
```

```python
snowballsstemmer=SnowballStemmer('english')
```

```python
for word in words: 
    print(word+"------->"+snowballsstemmer.stem(word))
```
