```python
d1 = "The food is good"
d2 = "The food is bad"
d3 = "Pizza is Amazing"

# Fixed vocabulary list
unique_words = ['The', 'Amazing', 'food', 'bad', 'good', 'Pizza', 'is', 'nice']

# One-hot encoding function
def one_hot_encode(sentence,vocab):
    words = sentence.split()
    encode = []
    for word in words:
        vector = [1 if word == v else 0 for v in vocab]
        encode.append(vector)
    return encode

# List of all sentences
sentences = [d1, d2, d3]

# Loop through each sentence and encode
for i, sentence in enumerate(sentences, 1):
    encoded = one_hot_encode(sentence, unique_words)
    print(f"Encoded d{i}:")
    for vector in encoded:
        print(vector)
    print()  # blank line between sentences
```
