```python
from collections import Counter
import pandas as pd
from nltk.corpus import stopwords
import nltk

nltk.download('stopwords')

# Sentences
sentence1 = "He is a very good boy and belong to good family."
sentence2 = "He is a very good girl and belong to good society."
sentence3 = "Boy and girl are good"

sentences = [sentence1, sentence2, sentence3]

stop_words = set(stopwords.words('english'))

# Process all sentences
all_words = " ".join(sentences).lower().split()
filter_words = [w for w in all_words if w not in stop_words]

# Vocabulary (words to vectorize) from unique filtered words
vocab = list(set(filter_words))

# Print vocabulary as a list
print("[", end="")
for i in vocab:
    print(f"'{i}'", end=", ")
print("]")

# Create vectors
vectors = []
for s in sentences:
    words = s.lower().split()
    counter = Counter(words)
    vector = [counter[word] for word in vocab]  # use the vocab variable
    vectors.append(vector)

# Create DataFrame
df = pd.DataFrame(vectors, columns=vocab, index=['Sentence1', 'Sentence2', 'Sentence3'])

# Display
print("\nVector Table:")
print(df)
```
