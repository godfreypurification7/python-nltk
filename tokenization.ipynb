```python
# https://www.nltk.org/_modules/nltk.html
# https://spacy.io/api/doc
# https://www.youtube.com/watch?v=ENLEjGozrio 45
# https://github.com/krishnaik06/The-Grand-Complete-Data-Science-Materials
```

```python
corpus= '''
hello welcome, to Krish Naik's.
please do watch the course! to be an expert.
'''
```

```python
print(corpus)
```

```python
##tokenization
## sentence --> paragraph
from nltk.tokenize import sent_tokenize
```

```python
documents=sent_tokenize(corpus)
```

```python
type(documents)
```

```python
for sentence in documents: 
    print(sentence)
```

```python
#tokenization technique
##paragraph --> words
from nltk.tokenize import word_tokenize 
```

```python
word_tokenize(corpus)
```

```python
for sentence in documents: 
    print(word_tokenize(sentence))
```

```python
from nltk.tokenize import wordpunct_tokenize
```

```python
wordpunct_tokenize(corpus)
```

```python
from nltk.tokenize import TreebankWordTokenizer
```

```python
tokenizer=TreebankWordTokenizer()
```

```python
tokenizer.tokenize(corpus)
```

```python
from nltk.tokenize import TreebankWordDetokenizer
```

```python
tokenizer=TreebankWordDetokenizer()
```

```python
tokenizer.tokenize(corpus)
```
