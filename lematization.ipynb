```python
from nltk.stem import WordNetLemmatizer 
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
'''
POS-Noun-n
verb-v
adjecctive-a
adverb-r
'''
lemmatizer.lemmatize('going',pos='v')
```

```python
lemmatizer.lemmatize('going',pos='v')
```

```python
for word in words: 
    print(word+"------>"+lemmatizer.lemmatize(word,pos='n'))
```
